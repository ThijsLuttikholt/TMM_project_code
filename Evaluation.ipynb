# Evaluation code for all research questions

### General code, usable for multiple questions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import ttest_rel
```

### Research question 1:

```python
data1 = np.load('processed_rq1.npy')
print(data1.shape)
```

###### Question 1a:

```python
plt.scatter(data1[:,1],data1[:,2])
plt.xlabel('average spoiler frequency')
plt.ylabel('average rating')
plt.title('Average spoiler frequency vs average rating per book')
```

```python
coef1a, p_val1a = pearsonr(data1[:,1], data1[:,2])
print(coef1a, p_val1a)
```

##### Question 1b:

```python
plt.scatter(data1[:,1],data1[:,3])
plt.xlabel('average spoiler frequency')
plt.ylabel('amount of reviews')
plt.title('Average spoiler frequency vs amount of reviews per book')
```

```python
coef1b, p_val1b = pearsonr(data1[:,1], data1[:,3])
print(coef1b, p_val1b)
```

### Question 2:

```python
data2 = np.load('processed_rq2.npy')
print(data2.shape)
```

##### Question 2a:

```python
plt.bar(['spoiler', 'non-spoiler'], [np.mean(data2[:,0]),np.mean(data2[:,2])])
plt.ylabel('average length in amount of words')
plt.title('average length in amount of words for spoiler vs non-spoiler')
```

```python
t_stat2a, p_val2a = ttest_rel(data2[:,0],data2[:,2])
print(t_stat2a, p_val2a)
```

##### Question 2b:

```python
plt.bar(['spoiler', 'non-spoiler'], [np.mean(data2[:,1]),np.mean(data2[:,3])])
plt.ylabel('average percentage of emotion-like words')
plt.title('average percentage of emotion-like words for spoiler vs non-spoiler')
```

```python
t_stat2b, p_val2b = ttest_rel(data2[:,1],data2[:,3])
print(t_stat2b, p_val2b)
```
